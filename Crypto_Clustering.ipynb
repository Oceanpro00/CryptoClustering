```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[df_market_data.columns])
```

```python
# Create a DataFrame with the scaled data
market_data_scaled_df = pd.DataFrame(market_data_scaled, columns=df_market_data.columns)

# Copy the crypto names from the original DataFrame
market_data_scaled_df["cryptos"] = df_market_data.index

# Set the coin_id column as index
market_data_scaled_df.set_index("cryptos", inplace=True)

# Display the scaled DataFrame
market_data_scaled_df
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))

# Create an empty list to store the inertia values
inertia = []
```

```python
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
	elbow_model = KMeans(n_clusters=i, random_state=1)
	elbow_model.fit(market_data_scaled_df)
	inertia.append(elbow_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
	"k": k,
	"inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_df = pd.DataFrame(elbow_data)
elbow_data_df
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_data_df.hvplot.line(
	x = "k",
	y = "inertia",
	title = "Elbow Chart for Crypto Market Data",
	xticks = k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Looking at the Elbow Chart for the Crypto Market Data, we can see that k = 4 will with a high certainty give us the optimal number of clusters, as the line shows there to be minimal representational benefit by moving any further down k.

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
k_model.fit(market_data_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
k_prediction = k_model.predict(market_data_scaled_df)

# Print the resulting array of cluster values.
k_prediction
```

```python
# Create a copy of the scaled DataFrame
market_data_clustered = market_data_scaled_df.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
market_data_clustered['k_clusters'] = k_prediction

# Display the copy of the scaled DataFrame
market_data_clustered
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
market_data_clustered.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by='k_clusters',
    hover_cols=['cryptos'],
    title = "K-Means Clustering of Cryptocurrencies"
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
market_data_pca = pca.fit_transform(market_data_scaled_df)

# View the scaled PCA data
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
display(pca.explained_variance_ratio_)
display(pca.explained_variance_ratio_.sum())
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The Total explained vaiance of the 3 principal components that we created equals out to **0.895** pointing to only about **90%** of the total data variance being represented by these 3 columns which honestly isnt great especially considering the data is on highly volatile crypto financial data.

```python
# Create a new DataFrame with the PCA data.
market_data_pca_df = pd.DataFrame(
    market_data_pca,
    columns=['PCA1','PCA2', 'PCA3'])

# Copy the crypto names from the original scaled DataFrame
market_data_pca_df['cryptos'] = market_data_scaled_df.index

# Set the coin_id column as index
market_data_pca_df.set_index('cryptos', inplace=True)

# Display the scaled PCA DataFrame
market_data_pca_df
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
pca_k = list(range(1,11))

# Create an empty list to store the inertia values
pca_inertia = []
```

```python
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in pca_k:
    model_pca = KMeans(n_clusters=i, random_state=1)
    model_pca.fit(market_data_pca_df)
    pca_inertia.append(model_pca.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {
    "k":pca_k,
    "inertia":pca_inertia
}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_data)

pca_elbow_df
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
pca_elbow_df.hvplot.line(
    x= 'k',
    y = 'inertia',
    xticks = pca_k,
    title = "Elbow Curve"
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Looking at the Elbow curve chart using the PCA Data, it is clear that the optimal value for k in this case would be 4 as there is little to no increase in the representation of the data's variation when evaluating further down the values of K.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, the Elbow Curve Chart constructed with or without using PCA showed little to no difference between one another. Visually said the charts look identical which points to the efficiency and effectivity in using PCA to find the optimal k values especially for larger data sets.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
pca_kmodel = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
pca_kmodel.fit(market_data_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_k_predictions = pca_kmodel.predict(market_data_pca_df)

# Print the resulting array of cluster values.
pca_k_predictions
```

```python
# Create a copy of the scaled PCA DataFrame
market_data_pca_clustered = market_data_pca_df.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
market_data_pca_clustered['pca_clusters'] = pca_k_predictions

# Display the copy of the scaled PCA DataFrame
market_data_pca_clustered.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
market_data_pca_clustered.hvplot.scatter(
    x = "PCA1",
    y = "PCA2",
    by = "pca_clusters",
    hover_cols=["cryptos"],
    title = "PCA Clustering Results"
)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
plot1 = elbow_data_df.hvplot.line(x='k', y='inertia', xticks=k, label="K-Means Clustering")
plot2 = pca_elbow_df.hvplot.line(x='k', y='inertia', xticks=k, label="PCA Clustering")

composite_plot1 = plot1 + plot2

composite_plot1
```

```python
# Composite plot to contrast the Elbow curves
plot1 = elbow_data_df.hvplot.line(x='k', y='inertia', xticks=k, label="regular plot")
plot2 = pca_elbow_df.hvplot.line(x='k', y='inertia', xticks=k, label="pca plot")

composite_plot1 = plot1 * plot2.opts(title='Regular vs PCA Data Elbow Method Comparison')

composite_plot1
```

```python
plot3 = market_data_clustered.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by='k_clusters',
    hover_cols=["Crypto Name", "cryptos"],
    label="K-Means Clustering",
    s=80,  # marker size
    cmap='Category10'
)

plot4 = market_data_pca_clustered.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="pca_clusters",
    hover_cols=["Crypto Name", "cryptos"],
    label="PCA Clustering",
    s=80,  # marker size
    cmap='Set1'  # contrasting color map
)

composite_plot3 = plot3 + plot4

composite_plot3
```

  * **Answer:** While the PCA does an excellent job of capturing the overall variability in the data and closely mimics the original structure, you can also see from the Elbow Method comparison plot that using fewer features leads to a noticeable drop in total inertia  which lines up with the PCAs explained variance ratios. One thing I found interesting, and became much clearer when looking at the PCA plot data, was that outliers stood out more in the reduced-feature clustering.
